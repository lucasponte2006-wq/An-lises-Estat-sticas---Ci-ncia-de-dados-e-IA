```r
# Instalar pacotes se necessário
# install.packages("ggplot2")

library(ggplot2)
```

```r
# Histograma - Salário
ggplot(dados, aes(x=Salario)) +
  geom_histogram(binwidth=500, fill="skyblue", color="black") +
  ggtitle("Distribuição de Salários")
```

```r
# Boxplot - Salário
ggplot(dados, aes(y=Salario)) +
  geom_boxplot(fill="orange") +
  ggtitle("Boxplot - Salário")
```

#Análise Salário
Os salários variam bastante, indo de R$ 800 até R$ 7.000.
A média ficou em torno de R$ 2.850, mas a mediana (R$ 2.500) mostra que boa parte ganha menos que a média — ou seja, temos alguns salários mais altos puxando esse valor para cima.

Dois funcionários (IDs 11 e 29) aparecem com R$ 7.000, que são considerados outliers. Esses casos merecem atenção para confirmar se são realmente salários diferenciados ou erros de registro.

No geral, a distribuição é bem desigual, com a maior parte dos funcionários entre R$ 1.800 e R$ 3.500, e poucos com salários muito altos.
