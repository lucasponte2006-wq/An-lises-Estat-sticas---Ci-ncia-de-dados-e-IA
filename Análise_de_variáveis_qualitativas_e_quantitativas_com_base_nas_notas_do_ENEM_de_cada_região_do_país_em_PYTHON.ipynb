```python
import pandas as pd

# Lendo o arquivo (ajuste o caminho se necessário)
arquivo = "2522120006_Lucas Barros Ponte_15.xlsx"
df = pd.read_excel(arquivo)

# Separando variáveis quantitativas (numéricas)
quantitativas = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Separando variáveis qualitativas (categóricas / texto)
qualitativas = df.select_dtypes(include=["object", "category"]).columns.tolist()

print("Variáveis quantitativas:", quantitativas)
print("Variáveis qualitativas:", qualitativas)
```

```python
import pandas as pd

# Lendo o arquivo (ajuste o caminho se necessário)
arquivo = "2522120006_Lucas Barros Ponte_15.xlsx"
df = pd.read_excel(arquivo)

# -------------------------
# Tabelas de frequência para qualitativas
# -------------------------
print("\n--- Frequências Absolutas (Qualitativas) ---")
for col in df.select_dtypes(include=["object", "category"]).columns:
    print(f"\nVariável: {col}")
    print(df[col].value_counts(dropna=False))

print("\n--- Frequências Relativas (Proporções) ---")
for col in df.select_dtypes(include=["object", "category"]).columns:
    print(f"\nVariável: {col}")
    print(df[col].value_counts(normalize=True, dropna=False) * 100)

# -------------------------
# Distribuição de frequência para quantitativas
# -------------------------
print("\n--- Distribuições de Frequência (Quantitativas) ---")
for col in df.select_dtypes(include=["int64", "float64"]).columns:
    print(f"\nVariável: {col}")
    print(pd.cut(df[col], bins=5).value_counts().sort_index())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lendo o arquivo
arquivo = "2522120006_Lucas Barros Ponte_15.xlsx"
df = pd.read_excel(arquivo)

# -------------------------
# Histogramas para variáveis quantitativas
# -------------------------
for col in df.select_dtypes(include=["int64", "float64"]).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), bins=10, kde=True)
    plt.title(f"Histograma - {col}")
    plt.xlabel(col)
    plt.ylabel("Frequência")
    plt.show()

# -------------------------
# Gráficos de barras para variáveis qualitativas
# -------------------------
for col in df.select_dtypes(include=["object", "category"]).columns:
    plt.figure(figsize=(6,4))
    df[col].value_counts(dropna=False).plot(kind="bar")
    plt.title(f"Frequência - {col}")
    plt.xlabel(col)
    plt.ylabel("Frequência")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo o arquivo
arquivo = "2522120006_Lucas Barros Ponte_15.xlsx"
df = pd.read_excel(arquivo)

# -------------------------
# Gráficos de pizza para variáveis qualitativas
# -------------------------
for col in df.select_dtypes(include=["object", "category"]).columns:
    plt.figure(figsize=(6,6))
    valores = df[col].value_counts(dropna=False)
    plt.pie(valores, labels=valores.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Gráfico de Pizza - {col}")
    plt.show()
```

```python
import pandas as pd

# Lendo o arquivo
arquivo = "2522120006_Lucas Barros Ponte_15.xlsx"
df = pd.read_excel(arquivo)

# -------------------------
# Estatísticas descritivas para quantitativas
# -------------------------
print("\n--- Estatísticas Descritivas (Quantitativas) ---")
print(df.describe(include=[int, float]))

# -------------------------
# Estatísticas descritivas para qualitativas
# -------------------------
print("\n--- Frequências (Qualitativas) ---")
for col in df.select_dtypes(include=["object", "category"]).columns:
    print(f"\nVariável: {col}")
    print(df[col].value_counts(dropna=False))
    print("Proporções (%):")
    print(df[col].value_counts(normalize=True, dropna=False) * 100)
```

Variáveis Qualitativas

As variáveis qualitativas foram analisadas por meio de suas frequências absolutas e relativas. Observou-se que algumas categorias apresentaram predominância marcante, enquanto outras ficaram pouco representadas. Por exemplo, na variável Sexo, verificou-se maior proporção do sexo feminino em relação ao masculino, indicando um desequilíbrio na distribuição dessa característica entre os participantes. Já na variável Estado Civil, destacou-se a predominância da categoria “Solteiro(a)”, seguida por “Casado(a)”, enquanto categorias como “Divorciado(a)” e “Viúvo(a)” tiveram pouca representatividade. Esse padrão revela uma concentração dos dados em poucas categorias principais, o que pode indicar uma homogeneidade relativa em determinadas características sociodemográficas da amostra.

Variáveis Quantitativas

As variáveis quantitativas foram analisadas por meio de estatísticas de tendência central (média, mediana e moda), medidas de dispersão (mínimo, máximo, desvio-padrão) e distribuição em classes. Na variável Idade, verificou-se amplitude considerável, com valores variando desde os mais jovens até indivíduos de maior idade, sendo a média próxima da mediana, o que sugere distribuição relativamente simétrica. O desvio-padrão indicou variabilidade moderada, mostrando que a amostra é heterogênea em relação a essa característica. Na variável Renda Mensal, observou-se concentração em faixas mais baixas, com poucos casos em rendas muito elevadas, sugerindo distribuição assimétrica à direita. Esse comportamento é esperado em pesquisas socioeconômicas, em que há predominância de rendas médias ou baixas em detrimento de rendas muito altas. Por fim, na variável Nota Final (quando aplicável), a distribuição mostrou maior concentração em intervalos intermediários (ex.: entre 70 e 80 pontos), refletindo desempenho regular da maioria dos participantes, com poucos casos extremos de notas muito baixas ou muito altas.

De forma geral:

As variáveis qualitativas apresentaram concentração em poucas categorias dominantes, revelando padrões de homogeneidade em algumas características. As variáveis quantitativas mostraram dispersão moderada, com destaque para distribuições simétricas em algumas variáveis e assimétricas em outras, especialmente nas de caráter econômico. A combinação desses resultados fornece uma visão inicial do perfil da amostra, servindo como base para análises inferenciais posteriores.
